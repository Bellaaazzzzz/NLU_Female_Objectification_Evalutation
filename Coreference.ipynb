### Import & Data Loading

```python
import allennlp_models.tagging
import spacy
from spacy import displacy
from spacy import tokenizer
from collections import Counter
from allennlp.predictors.predictor import Predictor
nlp = spacy.load('en_core_web_sm')
nlp.max_length = 1200000

from tqdm.notebook import tqdm
```

```python
with open('lolita.txt') as f:
    text_whole = f.read().replace('\n', ' ')

def chunkstring(string, length):
    return (string[0+i:length+i] for i in range(0, len(string), length))
```

```python
text_split = list(chunkstring(text_whole, 1000))
print(len(text_split))
text_0 = text_split[0:331]
text_1 = text_split[331:500]
text_2 = text_split[500:]
```

### Coreference

```python
predictor_cor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/coref-spanbert-large-2021.03.10.tar.gz")
```

```python
# corefernce. coref is hoped to be resolved text
coref_res = []
for segment in tqdm(text_0):
    coref_res.append(predictor_cor.coref_resolved(document=segment))
coref_res
text_file = open("Lolita_0.txt", "w")
text_file.write(' '.join(coref_res))
text_file.close()
```

```python
coref_res = []
for segment in tqdm(text_1):
    coref_res.append(predictor_cor.coref_resolved(document=segment))
coref_res
text_file = open("Output_1_LO.txt", "w")
text_file.write(' '.join(coref_res))
text_file.close()
```

```python
coref_res = []
for segment in tqdm(text_2):
    coref_res.append(predictor_cor.coref_resolved(document=segment))
coref_res
text_file = open("Output_2_LO.txt", "w")
text_file.write(' '.join(coref_res))
text_file.close()
```
